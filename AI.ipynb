```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
weight = ctrl.Antecedent(np.arange(0,10.1,0.1),'weight')
time =  ctrl.Antecedent(np.arange(0,10.1,0.1),'time')
power = ctrl.Consequent(np.arange(0,25.1,0.1),'power')
```

```python
#khối lượng gạo
weight['low'] = fuzz.trimf(weight.universe,[200,700,1000])
weight['vlow'] = fuzz.trimf(weight.universe,[200,200,700])
weight['normal'] = fuzz.trimf(weight.universe,[700,1000,1500])
weight['much'] = fuzz.trimf(weight.universe,[1000,1500,2000])
weight['vmuch'] = fuzz.trimf(weight.universe,[1000,2000,2000])

#thời gian
time['vslow'] = fuzz.trimf(time.universe,[10,10,50])
time['slow'] = fuzz.trimf(time.universe,[10,50,100])
time['normal'] = fuzz.trimf(time.universe,[50,100,130])
time['fast'] = fuzz.trimf(time.universe,[100,130,180])
time['vfast'] = fuzz.trimf(time.universe,[130,180,180])
#công suất
power['vlow'] = fuzz.trimf(power.universe,[0,00,25])
power['low'] = fuzz.trimf(power.universe,[0,25,50])
power['average'] = fuzz.trimf(power.universe,[25,50,75])
power['height'] = fuzz.trimf(power.universe,[50,75,100])
power['vheight'] = fuzz.trimf(power.universe,[75,75,100])
```
